```python
import torch

# Initialize a small weight matrix with values less than 1
weight = torch.full((3, 3), 0.5)

# Simulate forward pass through multiple layers (for simplicity)
hidden = torch.randn(3, 1)  # Random input vector

# Simulate passing through 10 layers in a deep network
for i in range(10):
    hidden = weight @ hidden  # Matrix multiplication to simulate layer operation

print("Output after 10 layers:", hidden)
```

```python
# Initialize a weight matrix with values greater than 1
weight = torch.full((3, 3), 1.5)  # Change the value to be greater than 1

# Simulate forward pass through multiple layers (for simplicity)
hidden = torch.randn(3, 1)  # Random input vector

# Simulate passing through 10 layers in a deep network
for i in range(100):
    hidden = weight @ hidden  # Matrix multiplication to simulate layer operation

print("Output after 10 layers:", hidden)
```
